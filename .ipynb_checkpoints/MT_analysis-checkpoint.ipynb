```python
import numpy as np
import pickle
from src.load_results import *
from src.plot_tools import *
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt

%matplotlib inline

plot_option = "print_and_save" ## "print" or "save" or "print_and_save"
```

```python
list_type_clf = ['SVM', 'KernelRidge']
list_type_of_CV = ['MT_Nested5foldCV', 'MT_5foldCV', 'MT_10foldCV', "MT_LOOCV"]
list_type_of_fold = ['S1', 'S2', 'S3', 'S4']

list_type_of_CV = ['MT_Nested5foldCV', 'MT_5foldCV', 'MT_LOOCV']
list_type_clf = ['SVM', 'KernelRidge']
list_type_of_fold = ['S1', 'S2', 'S3', 'S4']

dico_auc, dico_auc_stdev, dico_aupr, dico_aupr_stdev = \
    load_results(list_type_of_CV, list_type_clf, list_type_of_fold)

```

```python
for type_clf in list_type_clf:
    print(type_clf)
    dico_plot = {'pp':None, 'f_out':None}
    if plot_option=="save" or plot_option=="print_and_save":
        dico_plot['pp'] = PdfPages('saved_results/plots/MT/'+type_clf+'_score_depending_on_CV.pdf')
        dico_plot['f_out'] = open('saved_results/plots/MT/'+type_clf+'_score_depending_on_CV.txt', 'w')
    list_score1_temp = []
    list_score2_temp = []
    list_score1_stdev_temp = []
    list_score2_stdev_temp = []
    type_of_fold = 'S1'
    if type_clf=='SVM':
        list_name = []
        for type_of_CV in list_type_of_CV:
            list_score1_temp.append(dico_auc[type_of_CV][type_clf][type_of_fold])
            list_score1_stdev_temp.append(dico_auc_stdev[type_of_CV][type_clf][type_of_fold])
            list_score2_temp.append(dico_aupr[type_of_CV][type_clf][type_of_fold])
            list_score2_stdev_temp.append(dico_aupr_stdev[type_of_CV][type_clf][type_of_fold])
            if type_of_CV=='MT_Nested5foldCV':
                list_name.append('Nested\n 5-fold CV')
            elif type_of_CV=='MT_5foldCV':
                list_name.append('5-fold CV')
            elif type_of_CV=='MT_LOOCV':
                list_name.append('LOO CV')
        legend, xtickNames = plot_score_histo('', list_score1_temp, list_score2_temp, list_score1_stdev_temp, list_score2_stdev_temp, 'ROC-AUC', 'AUPR', '', list_name, plot_option, **dico_plot)
        
    elif type_clf=='KernelRidge':
        for type_of_CV in ['MT_Nested5foldCV', 'MT_5foldCV']:
            list_score1_temp.append(dico_auc[type_of_CV][type_clf][type_of_fold])
            list_score1_stdev_temp.append(dico_auc_stdev[type_of_CV][type_clf][type_of_fold])
            list_score2_temp.append(dico_aupr[type_of_CV][type_clf][type_of_fold])
            list_score2_stdev_temp.append(dico_aupr_stdev[type_of_CV][type_clf][type_of_fold])
        legend, xtickNames = plot_score_histo('', list_score1_temp, list_score2_temp, list_score1_stdev_temp, list_score2_stdev_temp, 'ROC-AUC', 'AUPR', '', ['MT_Nested5foldCV', 'MT_5foldCV'], plot_option, **dico_plot)
        #plt.setp(xtickNames, rotation=45)
    if plot_option=="save" or plot_option=="print_and_save":
        plt.savefig(dico_plot['pp'], format='pdf', bbox_extra_artists=(legend,), bbox_inches='tight')
        dico_plot['pp'].close()
        dico_plot['f_out'].close()
    if plot_option=="print" or plot_option=="print_and_save":
        plt.show()
    plt.clf()
```
